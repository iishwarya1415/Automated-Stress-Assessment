```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('STRESS.csv')
df.head()
```

```python
df.columns
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x='stress level',data=df)
```

```python
plt.figure(figsize=(15,5))

plt.subplot(1,2,1)
plt.hist(df['limb movement'],color='red')

plt.subplot(1,2,2)
plt.hist(df['blood oxygen'],color='blue')
```

```python
df['body temperature'].plot(kind='density')
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.heatmap(df.corr(),annot = True, fmt='0.2%',cmap = 'autumn',ax=ax)
```

```python
def plot(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(9,9), autopct='%1.2f%%', fontsize = 10)
    ax.set_title(variable + ' \n', fontsize = 10)
    return np.round(dataframe_pie/df.shape[0]*100,2)

plot(df, 'stress level')
```
